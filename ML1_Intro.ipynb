```python
# 2. Creaci√≥n de un Dataset de Negocio (Simulado)
# Imaginemos que tenemos los datos de inversi√≥n en publicidad y ventas de 10 meses
data = {
    'Inversion_Publicidad': [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000],
    'Ventas_Reales': [1500, 2500, 2800, 4200, 5100, 5800, 7200, 7900, 9100, 10500]
}
df = pd.DataFrame(data)
```

```python
print("--- NUESTROS DATOS DE LOS √öLTIMOS 10 MESES ---")
print(df)
```

```python
# 3. ESTAD√çSTICA DESCRIPTIVA (Lo que ya conocen)
print("\n--- RESUMEN ESTAD√çSTICO ---")
print(f"Venta promedio: {df['Ventas_Reales'].mean():.2f} ‚Ç¨")
print(f"Inversi√≥n total: {df['Inversion_Publicidad'].sum()} ‚Ç¨")
print(df.describe())
```

```python
# 4. VISUALIZACI√ìN (Mirando el pasado)
plt.figure(figsize=(8,5))
plt.scatter(df['Inversion_Publicidad'], df['Ventas_Reales'], color='blue', label='Datos Hist√≥ricos')
plt.title('Relaci√≥n Publicidad vs Ventas')
plt.xlabel('Inversi√≥n en Publicidad (‚Ç¨)')
plt.ylabel('Ventas (‚Ç¨)')
plt.grid(True)
plt.show()
```

```python
# 5. MACHINE LEARNING: PREDICCI√ìN (El futuro)
# Vamos a "entrenar" un modelo muy simple (Regresi√≥n Lineal)
X = df[['Inversion_Publicidad']] # Variable predictora
y = df['Ventas_Reales']           # Lo que queremos predecir

modelo = LinearRegression()
modelo.fit(X, y) # El modelo "aprende" la relaci√≥n
```

```python
# 6. LA PREGUNTA DE NEGOCIO:
# "¬øSi el mes que viene invertimos 12.000‚Ç¨, cu√°nto venderemos?"
nueva_inversion = [[12000]]
prediccion = modelo.predict(nueva_inversion)

print(f"\n--- PREDICCI√ìN ESTRAT√âGICA ---")
print(f"Si invertimos 12.000‚Ç¨, el modelo predice una venta de: {prediccion[0]:.2f} ‚Ç¨")
```
