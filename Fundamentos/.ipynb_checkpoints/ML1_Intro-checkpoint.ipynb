## PARTE A ‚Äì An√°lisis del Pasado (Estad√≠stica Descriptiva)

```python
#ruta_archivo = "/content/Advertising.csv" # Ejemplo ruta en Colab
ruta_archivo = "C:/tlsi/UCV_ML_Course/datasets/Advertising.csv" # Ejemplo ruta en local
df = pd.read_csv(ruta_archivo)
```

```python
print("--- NUESTROS DATOS ---")
df.drop(columns=['Unnamed: 0'], inplace=True)
print(df)
```

```python

print("\n RESUMEN ESTAD√çSTICO")
print("----")
print(f"Venta promedio: {df['Sales'].mean():.2f} M$")
print(df.describe())
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['TV'], df['Sales'], color='blue', label='Datos Hist√≥ricos')
plt.title('Relaci√≥n Gasto en TV vs Ventas')
plt.xlabel('Inversi√≥n en Publicidad en TV ($)')
plt.ylabel('Ventas (M$)')
plt.grid(True)
plt.show()
```

## PARTE B ‚Äì Anticipando el Futuro (Machine Learning)

```python
# 5. MACHINE LEARNING: PREDICCI√ìN (El futuro)
# Vamos a "entrenar" un modelo muy simple (Regresi√≥n Lineal)
X = df[['TV']] # Variable predictora
y = df['Sales']           # Lo que queremos predecir

modelo = LinearRegression()
modelo.fit(X, y) # El modelo "aprende" la relaci√≥n
```

```python
# 6. LA PREGUNTA DE NEGOCIO:
# "¬øSi el mes que viene invertimos $400, cu√°nto venderemos?"
nueva_inversion_TV = [[400]]
prediccion = modelo.predict(nueva_inversion_TV)

print(f"\n--- PREDICCI√ìN ESTRAT√âGICA ---")
print(f"Si invertimos $400 en publicidad en TV, el modelo predice una venta de: {prediccion[0]:.2f} M$")

print("El modelo no memoriza los datos, aprende una regla general")
print(f"Peso de la variable predictora: {modelo.coef_[0]}")
print(f"Valor del intercepto (bias): {modelo.intercept_}")
```

```python
# Crear valores para la l√≠nea de regresi√≥n
X_linea = np.array(df['TV']).reshape(-1, 1)
y_linea = modelo.predict(X_linea)

# Gr√°fico
plt.figure(figsize=(8,5))

# Datos reales
plt.scatter(df['TV'], df['Sales'],
            label='Datos Hist√≥ricos', alpha=0.8)

# L√≠nea de regresi√≥n
plt.plot(df['TV'], y_linea,
         label='Modelo de Regresi√≥n', linewidth=2)

# Punto futuro
plt.scatter(400, prediccion[0],
            marker='X', s=120, label='Predicci√≥n Futura')

# Etiquetas y formato
plt.title('Publicidad vs Ventas: Pasado y Futuro')
plt.xlabel('Inversi√≥n en Publicidad en TV ($)')
plt.ylabel('Ventas (M$)')
plt.legend()
plt.grid(True)

plt.show()
```
